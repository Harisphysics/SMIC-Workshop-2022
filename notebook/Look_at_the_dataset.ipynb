Creating Pipeline

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
data_num = data.select_dtypes(exclude=["object"])
data_cat = data.select_dtypes(include=["object"])
```

```python
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("std_scaler", StandardScaler()),
])
```

```python
num_pipeline.fit_transform(data_num)
```

```python
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, data_num.columns),
    ("cat", OneHotEncoder(), data_cat.columns)
])
```

```python
pd.DataFrame(full_pipeline.fit_transform(data))
```

```python
def transformer(data):
  data_num = data.select_dtypes(exclude=["object"])
  data_cat = data.select_dtypes(include=["object"])
  num_pipeline = Pipeline([("imputer", SimpleImputer(strategy="mean")),
                           ("std_scaler", StandardScaler())])
  full_pipeline = ColumnTransformer([("num", num_pipeline, data_num.columns),
                                     ("cat", OneHotEncoder(), data_cat.columns)])
  df = full_pipeline.fit_transform(data)
  return df
```

```python
transformer(data)
```
